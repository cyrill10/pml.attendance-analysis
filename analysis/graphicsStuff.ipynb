```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
attendance_df = pd.read_csv('../data/matches_prep.csv', sep=',')

```

```python
basel_home = attendance_df.loc[attendance_df['hometeam'] == 'FC Basel']
zurich_home = attendance_df.loc[attendance_df['hometeam'] == 'FC Zürich']
```

```python
basel_avg = (basel_home.groupby(['season'])).mean()['attendance']
zurich_avg = (zurich_home.groupby(['season'])).mean()['attendance']
zurich_max = (zurich_home.groupby(['season'])).max()['attendance']
```

```python
zurich_avg
```

```python
zurich_basel = zurich_home.loc[zurich_home['awayteam'] == 'FC Basel']
basel_zurich = basel_home.loc[basel_home['awayteam'] == 'FC Zürich']
```

```python
basel_zurich_avg = (basel_zurich.groupby(['season'])).mean()['attendance']
zurich_basel_avg = (zurich_basel.groupby(['season'])).mean()['attendance']
```

```python
basel_zurich_avg
```

```python
plt.figure(figsize=(12, 9))
plt.ylim(4000, 36000)
plt.xlim(2002, 2019)
plt.ylabel("Zuschauerschnitt")
for y in range(5000, 35000, 5000):
    plt.plot(range(2002, 2019), [y] * len(range(2002, 2019)), "--", lw=0.5, color="black", alpha=0.3)
plt.plot(basel_zurich_avg, marker="o", color="r", linestyle="-", label='FC Basel - FC Zürich')
plt.plot(basel_avg, marker="o", color="lightcoral", linestyle="-", label='Heimspiele FC Basel')
plt.plot(zurich_basel_avg, marker="o", color="b", linestyle="-", label='FC Zürich - FC Basel' )
plt.plot(zurich_avg, marker="o", color="lightblue", linestyle="-", label='Heimspiele FC Zürich')
plt.legend(loc="upper left")
plt.text(2002, 37000, "Einfluss von Derbys auf das durchschnittliche Zuschaueraufkommen", fontsize=17)
plt.text(2002, 1000, "Datenquelle: weltfussball.com"
       "\nAuthor: Cyrill Bader", fontsize=10)
plt.savefig('../graphs/derby.png')
plt.show()
plt.close()
```

```python
xamax_home = attendance_df.loc[attendance_df['hometeam'] == 'Neuchâtel Xamax FC']
xamax_max = (xamax_home.groupby(['season'])).max()['attendance']
xamax_max
```

```python
gc_home = attendance_df.loc[attendance_df['hometeam'] == 'Grasshopper Club Zürich']
gc_avg = (gc_home.groupby(['season'])).mean()['attendance']
gc_max = (gc_home.groupby(['season'])).max()['attendance']
gc_avg
```

```python
plt.figure(figsize=(12, 9))
plt.ylim(4001, 27000)
plt.xlim(2006, 2019)
plt.ylabel("Zuschauerzahlen")
for y in range(5000, 26000, 5000):
    plt.plot(range(2002, 2019), [y] * len(range(2002, 2019)), "--", lw=0.5, color="black", alpha=0.3)
plt.plot(zurich_max.loc[2007:2019], marker="o", color="lightblue", linestyle="-", label='Maximale Zuschauerzahl FC Zürich')
plt.plot(zurich_avg.loc[2007:2019], marker="o", color="blue", linestyle="-", label='Zuschauerschnitt FC Zürich')
plt.plot(gc_max.loc[2007:2019], marker="o", color="lightgreen", linestyle="-", label='Maximale Zuschauerzahl Grasshopper Club Zürich')
plt.plot(gc_avg.loc[2007:2019], marker="o", color="green", linestyle="-", label='Zuschauerschnitt Grasshopper Club Zürich')
plt.legend(loc="upper right")
plt.text(2006, 28000, "Zuschauervergleich GCZ und FCZ", fontsize=17)
plt.text(2006, 2000, "Datenquelle: weltfussball.com"
       "\nAuthor: Cyrill Bader", fontsize=10)
plt.savefig('../graphs/zurich.png')
plt.show()
plt.close()
```
